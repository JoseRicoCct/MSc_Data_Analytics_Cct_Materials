```python
import keras
from keras.models import Sequential
from keras.layers import Dense
import pandas as pd
import numpy as np
import sklearn
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('pima-indians-diabetes.csv',header=None)
```

```python
df.head()
```

```python
X = df.iloc[:,0:8] # selecting all columns except column 8 | Dependent variable
y = df.iloc[:,8] # selecting column 8 as our target variable | Independent variable
```

```python
X
```

```python
y
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
X
```

```python
model = Sequential()
model.add(Dense(10,input_dim=8,activation='relu')) # First layer
model.add(Dense(10,activation='relu')) # Second layer
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss="binary_crossentropy",optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(X,y,epochs=150,batch_size=10,verbose=2)
```

```python
scores = model.evaluate(X,y)
print('%s: %.2f%%' % (model.metrics_names[1],scores[1]*100))
```

```python
predictions = model.predict(X)
print(predictions)
```

```python
rounded = [round(x[0]) for x in predictions]
```

```python
print(rounded)
```

```python
import matplotlib.pyplot as plt
from sklearn import metrics

confusion_matrix = metrics.confusion_matrix(y,rounded)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_matrix,display_labels=[False,True])

cm_display.plot()
plt.show()
```
